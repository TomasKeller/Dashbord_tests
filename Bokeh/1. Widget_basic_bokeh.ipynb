```python
import pandas as pd
import numpy as np
# Create demo data
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
df['z'],df['zz'] = 1,1
df['w']=np.random.uniform(1,0,200)

df['grp'] = ['A' if x > 4 else 'B' if  x > 1 else 'C' for x in df.sum(axis=1)]
df['grpX'] = ['A' if x > 1 else 'B' if  x > .5 else 'C' for x in df.x]

df.head()

from bokeh.charts import Bar
#from bokeh.bkcharts import Bar
p = Bar(df, 'grpX', values='z')
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
plot = figure(plot_width=300, plot_height=300)
plot.hbar(y=[1, 2, 3], height=0.5, left=0, right=[1,2,3])
show(plot)
```

```python
from bokeh.charts import Scatter

p = Scatter(df, x='x', y='y', color='grp', legend='top_left')
show(p)
```
